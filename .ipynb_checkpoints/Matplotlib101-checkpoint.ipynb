```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 3
```

```python
numpyDizisi2
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"g--")
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"b*-")
```

```python
plt.subplot(1,2,1)
plt.plot(numpyDizisi1,numpyDizisi2,"r*-")
plt.subplot(1,2,2)# 1 sira olacak, 2 kolon olacak, su an ikinci grafigi ciziyorum.
plt.plot(numpyDizisi2,numpyDizisi1,"g--")
```

```python
myFigure = plt.figure()
figureAxes = myFigure.add_axes([0.1,0.1,0.6,0.6])
figureAxes.plot(numpyDizisi1,numpyDizisi2,"g")
figureAxes.set_xlabel("X Ekseni Veri Ismi")
figureAxes.set_ylabel("Y Ekseni Veri Ismi")
figureAxes.set_title("Grafik Basligi")
```

```python
figure2 = plt.figure()

eksen1 = figure2.add_axes([0.1,0.1,0.9,0.9])
eksen2 = figure2.add_axes([0.3,0.5,0.3,0.3])

eksen1.plot(numpyDizisi1,numpyDizisi2,"g")
eksen1.set_xlabel("X Ekseni")
eksen1.set_ylabel("Y Ekseni")
eksen1.set_title("Ana Grafik Baslik")

eksen2.plot(numpyDizisi2,numpyDizisi1,"b")
eksen2.set_xlabel("X Ekseni Kucuk")
eksen2.set_ylabel("Y Ekseni Kucuk")
eksen2.set_title("Kucuk Grafik Baslik")
```

```python
(benimFigure,benimEksenler) = plt.subplots(nrows=1, ncols=2)
type(benimEksenler)

for eksen in benimEksenler:
    eksen.plot(numpyDizisi1,numpyDizisi2,"b")
    eksen.set_xlabel("X Ekseni")
plt.tight_layout()
```

```python
#yeniFigure = plt.figure(figsize=(3,3))
#yeniFigure = plt.figure(dpi=100)
yeniFigure = plt.figure()


yeniEksen = yeniFigure.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyDizisi1,numpyDizisi1 ** 2,label = "numpyDizisi ** 2")
yeniEksen.plot(numpyDizisi1,numpyDizisi2,label = "numpyDizisi ** 3")

yeniEksen.legend(loc=2)
#yeniEksen.legend(loc=10)
```

```python
yeniFigure.savefig("benimfigur.png", dpi = 200)
```
