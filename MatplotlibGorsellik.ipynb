```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
numpyDizisi1 = np.linspace (0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 2
```

```python
numpyDizisi2
```

```python
(benimFigur, benimEksen) = plt.subplots()
benimEksen.plot(numpyDizisi1,numpyDizisi2,"#66FFB2", alpha = 0.5)
benimEksen.plot(numpyDizisi2,numpyDizisi1,"#F557A6")
```

```python
(yeniFigur, yeniEksen) = plt.subplots()

yeniEksen.plot(numpyDizisi1,numpyDizisi1 + 2, color="#66FFB2", linewidth = 6.0)
yeniEksen.plot(numpyDizisi1,numpyDizisi1 + 3,"#F557A6", linewidth = 4.0 )
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 4, color="#99004C", linestyle= "--")
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 5, color="#FF9933", linestyle= ":")
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 6, color="#0000FF", linestyle= "-.")

yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 7, color="#000000", linestyle= "--", marker="o", markersize = 8 , markerfacecolor = "red")
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 8, color="#000000", linestyle= "--", marker="+", markersize = 10 )
```
