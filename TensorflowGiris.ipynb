```python
from sklearn.model_selection import train_test_split
```

```python
#train_test_split
```

```python
# y = wx + b
# y -> label
# x -> feature (ozellik)
```

```python
y = dataFrame["Fiyat"].values
x = dataFrame[["BisikletOzellik1","BisikletOzellik2"]].values
x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.33,random_state=15)
```

## scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()

# 5 kac noron oldugunu belirtmek icin yazilir.
# Her biri bir tane Hidden Layer i temsil eder.

model.add(Dense(5,activation="relu"))
model.add(Dense(5,activation="relu"))
model.add(Dense(5,activation="relu"))

# Cikti noronu asagidaki gibi tercih edilir bu ornekte tek cikti oldugu icin 1 yazdim.

model.add(Dense(1))
model.compile(optimizer= "rmsprop", loss = "mse")
```

```python
model.fit(x_train,y_train,epochs=250)
```

```python
loss  = model.history.history["loss"]
```

```python
loss
```

```python
sbn.lineplot(x=range(len(loss)), y=loss)
```

```python
trainLoss = model.evaluate(x_train, y_train, verbose =1)
```

```python
trainLoss
```

```python
testLoss = model.evaluate(x_test, y_test, verbose=1)
```

```python
testLoss
```
